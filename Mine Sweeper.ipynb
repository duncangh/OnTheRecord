```python
import pandas as pd
import numpy as np
import numpy.random as npr
```

```python
def make_field(rows=9, cols=9, mines=10):
    total_cells = rows*cols
    non_mine_cells = total_cells - mines
    
    numbers = list([0] * non_mine_cells)
    bombs = list([1] * mines)
    numbers.extend(bombs)
    
    field = np.array(numbers)
    npr.shuffle(field)
    
    df = pd.DataFrame(field.reshape((9, 9)), columns=list('ABCDEFGHI'), index=list('ABCDEFGHI'))
    return df

def calc_local_mines():
    df = make_field()
    shifts = [df.shift(1) - df,
     df.shift(-1) - df,
    df.shift(1, axis=1) - df,
    df.shift(-1, axis=1) - df,
    df.shift(1).shift(axis=1) - df,
    df.shift(-1).shift(axis=1) - df,
    df.shift(1).shift(-1, axis=1) - df,
    df.shift(-1).shift(-1, axis=1) - df]
    main = df.copy()
    for shift in shifts:
        main += clean(shift)
    main[df == 1] = 'X'

    return main

def clean(df):
    return df.fillna(0).replace(-1, 0)
```

```python
def convert(x):
    try:
        return int(x)
    except:
        return x
df = calc_local_mines()
df.applymap(convert)
```
