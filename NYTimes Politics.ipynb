```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import nltk
from nltk.tag import pos_tag
```

```python
def nytimes_politics_article_links():
    url = 'https://www.nytimes.com/pages/politics/index.html'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, 'lxml')
    
    body = soup.find('div', {'id' : 'main'})
    links = body.find_all('a')
    articles = []
    for link in links:
        try:
            article_link = link['href']
            if 'us/politics' in article_link and 'interactive' not in article_link:
                articles.append(article_link)

        except:
            pass
    
    return articles

articles = nytimes_politics_article_links()
```

```python
def find(paragraph, ch='"'):
    idxs =  [i for i, ltr in enumerate(paragraph) if ltr == ch]
    start, stop = idxs[0::2], idxs[1::2]
    quotes = [paragraph[idx: stop[i] + 1] for i, idx in enumerate(start)]
    if len(quotes) > 0:
        return quotes

def new_york_times(url):
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, 'lxml')
    
    body = soup.find('article', {'id' : 'story'}).find_all('p', {'class' : 'story-body-text story-content'})
    text = ''''''
    quotes = []
    
    for paragraph in body:
        paragraph_text = paragraph.text.replace('“', '"').replace('”', '"').replace('."', '." ')
        text += paragraph_text + ' '
        results = find(paragraph_text)
        
        if results:
            quotes.extend(results)
        
    header = soup.find('article', {'id' : 'story'}).find('h1', {'class' : 'headline'}).text + '.txt'
    file_name = '/Users/duncangh/PycharmProjects/Articles/Files/%s' % header
    
    open(file_name, 'w').write(text)
    
    return header[:-4], quotes

```

```python
article_quotes = {}
for url in articles:

    title, quotes = new_york_times(url)
    article_quotes[title] = quotes
```
